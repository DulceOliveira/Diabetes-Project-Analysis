**Model**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, mean_squared_error

# Splitting the data into features and target

X = cleaned_diabetes.drop('Diabetes_Status', axis=1)  # Features
y = cleaned_diabetes['Diabetes_Status']  # Target

# Splitting the data into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

##### **XGBOOST Model**

```python
# XGBoost

from sklearn.metrics import accuracy_score, precision_score, recall_score, mean_squared_error
import xgboost as xgb

# Creating and train the XGBoost model

xgb_model = xgb.XGBClassifier(random_state=42)
xgb_model.fit(X_train, y_train)

# Making predictions

y_pred_xgb = xgb_model.predict(X_test)

# Evaluating the model

accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
precision_xgb = precision_score(y_test, y_pred_xgb)
recall_xgb = recall_score(y_test, y_pred_xgb)
mean_squared_error_xgb = mean_squared_error(y_test, y_pred_xgb)

# Printing the evaluation metrics

print("XGBoost:")
print(f"Accuracy: {accuracy_xgb}")
print(f"Precision: {precision_xgb}")
print(f"Recall: {recall_xgb}")
print(f"Mean Squared Error: {mean_squared_error_xgb}")
```

```python
# Saving the model to a pickle file

import pickle

# Save the model to a file
model_filename = 'diabetes_project_model.pkl'
with open(model_filename, 'wb') as model_file:
    pickle.dump(xgb_model, model_file)
```

```python
from sklearn.metrics import f1_score, roc_auc_score, roc_curve

# Calculating the F1-score

f1 = f1_score(y_test, y_pred_xgb)

# Calculating the ROC-AUC score

roc_auc = roc_auc_score(y_test, xgb_model.predict_proba(X_test)[:, 1])

# Printing the metrics

print(f"F1-Score: {f1}")
print(f"ROC-AUC: {roc_auc}")
```
